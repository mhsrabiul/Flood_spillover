```python
import numpy as np
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df_rabiul = pd.read_csv('FloodData.csv')
```

```python
df_rabiul.head()
```

```python
# Create a GeoDataFrame with Latitude and Longitude
gdf = gpd.GeoDataFrame(df_rabiul, geometry=gpd.points_from_xy(df_rabiul.longitude, df_rabiul.latitude))

# Plot suitability categories
fig, ax = plt.subplots(figsize=(10, 8))
gdf.plot(column='SUITABILITY', ax=ax, legend=True, cmap='viridis', alpha=0.5)
plt.title('Suitability Categories')
plt.xlabel('longitude')
plt.ylabel('latitude')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
gdf.plot(column='PROB_4BAND', ax=ax, legend=True, cmap='coolwarm', alpha=0.5)
plt.title('Flood Probability')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```
